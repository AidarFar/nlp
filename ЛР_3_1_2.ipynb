```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.model_selection import train_test_split

# Загрузка данных из датасета balanced_df
# Предположим, у вас есть столбец "text" с текстами и столбец "label" с метками классов
data = pd.read_csv('/content/drive/MyDrive/datasets/balanced_df.csv')
data
```

```python
texts = data['Review Text']
labels = data['Rating']
texts = texts.astype(str)
texts
```

```python

# Токенизация текста
max_words = 10000  # Максимальное количество слов в словаре
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
sequences
```

```python
# Паддинг последовательностей
max_sequence_length = 100  # Максимальная длина последовательности
data = pad_sequences(sequences, maxlen=max_sequence_length)

# Разделение данных на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Создание модели RNN
model = Sequential()
model.add(Embedding(max_words, 128, input_length=max_sequence_length))
model.add(LSTM(64))
model.add(Dense(1, activation='sigmoid'))

# Компиляция модели
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Оценка модели на тестовых данных
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Точность модели: {accuracy * 100:.2f}%')

# Можно использовать модель для предсказания новых текстов
# new_texts = ["Пример нового текста 1", "Пример нового текста 2"]
# new_sequences = tokenizer.texts_to_sequences(new_texts)
# new_data = pad_sequences(new_sequences, maxlen=max_sequence_length)
# predictions = model.predict(new_data)
```
