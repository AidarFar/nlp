```python
def crypto(message, sdvig):
    message_crypto = ""
    for i in message:
        message_crypto += chr(ord(i)+sdvig)
    return message_crypto
```

```python
def message_to_numbers(message):
    message_numbers = []
    for i in message:
        message_numbers.append(ord(i))
    return np.array(message_numbers)
```

```python
model_rnn = keras.Sequential()



model_rnn.add(LSTM(1024, activation='tanh', return_sequences=True))
model_rnn.add(LSTM(1024, activation='tanh', return_sequences=False))
model_rnn.add(Dense(1024, activation='tanh'))
model_rnn.add(Dense(1024, activation='tanh'))
model_rnn.add(Dense(1024, activation='tanh'))
model_rnn.add(Dense(100, activation='tanh'))

model_rnn.compile(loss='mse', optimizer=keras.optimizers.Adam(0.001))
```

```python
model_rnn.fit(X_train, y_train, epochs=100)
```

```python
def text_generate(text):
    vectors = []
    vectors.append(text.lower().split())
    for i in range(len(vectors[0])):
        try:
            vectors[0][i] = model_embedding[vectors[0][i]]
        except:
            vectors[0][i] = model_embedding['и']
    return np.array(vectors)

```

```python
text = text_generate('Я люблю котов и собак и даже возможно произведения федора')
```

```python
word_pred = model_rnn.predict(text)
```

```python
from random import randint
```

```python
def words_predict(text, number_of_words):
    text_for_work = text.split()
    for i in range(number_of_words):
        text = text_generate(" ".join(text_for_work[-10:]))
        text_for_work.append(model_embedding.similar_by_vector(model_rnn.predict(text)[0])[randint(0, 5)][0])
    return " ".join(text_for_work)
```
